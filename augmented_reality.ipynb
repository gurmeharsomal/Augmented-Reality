```python
def insert_cube(image, corners, pts):
    pts = np.int32(pts).reshape(-1,2)
    image = cv.drawContours(image, [pts[:4]],-1,(0,255,0),-3) #green base
    for i,j in zip(range(4),range(4,8)):
        image = cv.line(image, tuple(pts[i]), tuple(pts[j]),(255),3) #blue pillars
    image = cv.drawContours(image, [pts[4:]],-1,(0,0,255),3) #red top box
    return image
```

```python
def calibrating(twoDpts, threeDpts):
    
    Mat, s = np.zeros((len(threeDpts)*3, 11)), 1
    b, c = np.zeros((len(twoDpts)*3)), 0
    
    
    for i in range(len(threeDpts)): #amending the 3dworld coordinates
        Mat[c,:]= np.array([threeDpts[i][0], threeDpts[i][1], threeDpts[i][2], 1, 0, 0, 0, 0, 0, 0, 0])
        Mat[c + 1,:]= np.array([0, 0, 0, 0, threeDpts[i][0], threeDpts[i][1], threeDpts[i][2], 1, 0, 0, 0])
        Mat[c + 2,:]= np.array([0, 0, 0, 0, 0, 0, 0, 0, threeDpts[i][0], threeDpts[i][1], 1])
        c += 3    
    
    c = 0
    for k in range(len(twoDpts)): #amending the 2d coordinates
        b[c] = twoDpts[k][0]*s
        b[c + 1] = twoDpts[k][1]*s
        b[c + 2] = s - threeDpts[k][2]
        c += 3
        
    res = np.linalg.lstsq(Mat, b, rcond=None)[0]
    return np.hstack((res[:10], np.ones(1), h[10])).reshape(3,4)
```
